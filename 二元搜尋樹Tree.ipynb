```python
import sys
class Student:
  def __init__(self):
    self.id=0
    self.name=''
    self.score=0
    self.llink=None
    self.rlink=None
root=None

def insert_f():
  print('\n======INSERT DATA======')
  id=eval(input('Enter student id:'))
  name=input('Enter student name:')
  score=eval(input('Enter student score'))
  access(id,name,score)

def delete_f():
  if root==None:
    print('N0 student record!')
    return
  print('\n=====DELETE DATA======')
  id=eval(input('Enter student id:'))

  removing(id)

def modify_f():
  node=None
  if root==None:
    print('No student record!')
    return
  else:
    print('\n=====MODIFY DATA======')
    id=eval(input('Enter student id:'))

    node=search(id)
    if node==None:
      print('Student %d not found!'%(id))

    else:
      print('student id:',node.id)
      print('student name:',node.name)
      print('student score:',node.score)
      node.score=eval(input('Enter new score'))
      print('Student id %d has been modified'%(id))

def show_f():
  if root==None:
    print('Noe student record!')
    return
  print('\n%-10s %-15s %-6s'%('ID','NAME','SCORE'))
  inorder(root)

def access(id,name,score):
  global root
  node=None
  prev=None
  if search(id) !=None:
    print('Student id %d has existed!'%(id))
    return
  ptr=Student()
  ptr.id=id
  ptr.name=name
  ptr.score=score
  ptr.llink=None
  ptr.rlink=None
  if root==None:
    root=ptr
  else:
    node=root
    while node !=None:
      prev=node
      if ptr.id<node.id:
        node=node.llink
      else:
        node=node.rlink
    if ptr.id<prev.id:
      prev.llink=ptr
    else:
      prev.rlink=ptr

def removing(id):
  global root
  del_node=search(id)
  if del_node==None:
    print('Student id %d not found!'%(id))
    return
  if del_node.llink!=None or del_node.rlink !=None:
    del_node=replace(del_node)
  else:
    if del_node==root:
      root=None
    else:
      connect(del_node,'n')
  del_node=None
  print('Student id %d has been deleted!'%(id))

def replace(node):
  re_node=None
  re_node=search_re_r(node.rlink)
  if re_node==None:
    re_node=search_re_l(node.llink)
  if re_node.rlink !=None:
    connect(re_node,'r')
  elif re_node.llink !=None:
    connect(re_node,'l')
  else:
    connect(re_node,'n')
  node.id=re_node.id
  node.name=re_node.name
  node.score=re_node.score
  return re_node

def connect(node, link):
  parent=search_p(node)
  if node.id<parent.id:
    if link=='r':
      parent.llink=node.rlink
    elif link=='l':
      parent.llink=node.llink
    else:
      parent.llink=None
  else:
    if link=='r':
      parent.rlink=node.rlink
    elif link=='l':
      parent.rlink=node.llink
    else:
      parent.rlink=None
def inorder(node):
  if(node !=None):
    inorder(node.llink)
    print('%-10d %-15s %-3d'%(node.id,node.name,node.score))
    inorder(node.rlink)

def search(target):
  global root
  node= root
  while node !=None:
    if target==node.id:
      return node
    elif target < node.id:
      node=node.llink
    else:
      node=node.rlink
  return node

def search_re_r(node):
  re_node=node
  while re_node !=None and re_node.llink !=None:
    re_node=re_node.llink
  return re_node

def search_re_l(node):
  re_node=node
  while re_node !=None and re_node.rlink !=None:
    re_node=re_node.rlink
  return re_node

def search_p(node):
  global root
  parent=root
  while parent !=None:
    if node.id <parent.id:
      if node.id==parent.llink.id:
        return parent
      else:
        parent=parent.llinl
    elif node.id==parent.rlink.id:
      return parent
    else:
      parent=parent.rlink
  return None

def main():
  while True:
    print()
    print('********************')
    print('     1.insert     ')
    print('     2.delete     ')
    print('     3.modify     ')
    print('     4.show     ')
    print('     5.quit     ')

    try:
      option=input('Enter your choice:')
    except ValueError:
      print('Not a correct number')
      print('Try again\n')
    if option=='1':
      insert_f()
    elif option=='2':
      delete_f()
    elif option=='3':
      modify_f()
    elif option=='4':
      show_f()
    elif option=='5':
      break
    else:
      print('Wrong option!')
main()




```
