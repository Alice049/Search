# 使串列儲存heap資料，並利用last_index變數，記錄目前Heap最後一筆資料的index值

# 新增節點：
1.insert_f()函數：先要求使用者輸入新增節點的值，（若串列還有空間）再呼叫create()函數來建立新節點。
2.create()函數：先將資料加入Heap的最後一個節點後,再由下往上調整，使其符合Heap的定義
3.某節點的父節點的index值，該節點的Index值除以2,呼叫adjust_u()函數，使節點不斷往上調整，直到該節點所儲存的資料小於父節點。

# 刪除節點：
1.delect_f()函數：要求使用輸入要刪除的資料，先以search（）函數尋找該資料所在的節點後，再呼叫removes()函數從Heap串列中移除。
2.removes()函數：先將欲刪除資料與最後一個節點的資料交換，使用heap_tree[last_index]＝敘述將最後一筆資料的值指定為0,並透過last_index -=1將last_index減1
3.最後必須調整之前與最後一筆資料交換的節點。若該節點所儲存的資料小於父節點，則呼叫adjust_u()往上調整，否則呼叫adjust_d()往下調整，直到符合Heap定義為止。

# 輸出節點
輸出節點時可以選擇由小到大排列或由大到小排列
1.必須先產生一串列來儲存串列的值，以此新產生的Heap串列來儲存資料。
2.將最後一個節點的資料與第一個節點的資料進行交換，此時最後一個節點會儲存Heap的最大值，（最後一個節點index會遞減），該資料不會被更新，而第一個節點必須往下調整至符合Heap的定義，如此反覆進行至c_index值等於0。串列內資料的排序會由小到大排列。
3.串列中的資料若以堆疊的方式輸出，則可得到由大到小的排序資料;而以佇列方式輸出，則可得到由小到大的排序資料。






```python
import sys
MAX=100
heap_tree=100
heap_tree=[0]*MAX
last_index=0

def insert_f():
  if last_index >= MAX-1:
    print('\n Login members are more than %d !!'% MAX-1)
    print(' Please wait for a Minute...')

  id_temp=int(input('\n Please enter login ID number: '))
  create(id_temp)
  print(' login successfully!! ')

def delete_f():
  id_temp=0
  del_index=0
  if last_index<1:
    print('\n No menber to logout!!')
    print(' Please check again!!')
  else:
    id_temp=int(input('\n  Please enter logout ID number:'))
    del_index=search(id_temp)
    if del_index==0:
      print(' ID number not found!! ')
    else:
      removes(del_index)
      print(' ID number ',id_temp,'logout!!')

def display_f():
  if last_index <1:
    print('\n. No member to show!! \n')
  else:
    print()
    print('***************************')
    print('        <1>increase        ')
    print('        <2>decrease        ')
    print('***************************')
    while True:
      try:
        option=input('\n Please enter your option:')
      except ValueError:
        print()
        print('Not a correctly number.')
        print('Try again\n')
      if (option =='1' or option=='2'):
        break
    show(option)

def create(id_temp):
  global heap_tree
  global last_index

  last_index += 1
  heap_tree[last_index] = id_temp
  adjust_u(heap_tree,last_index)


def removes(index_temp):
  global heap_tree
  global last_index

  heap_tree[index_temp]=heap_tree[last_index]
  heap_tree[last_index]=0
  last_index -=1
  if last_index>1:
    if heap_tree[index_temp]>heap_tree[index_temp//2] and index_temp >1:
      adjust_u(heap_tree,index_temp)
    else:
      adjust_d(heap_tree,index_temp,last_index-1)

def show(op):
  heap_temp=[]
  tChar= ''

  heap_temp=[i for i in heap_tree]
  c_index=last_index-1
  while c_index>0:
    exchange(heap_temp,1,c_index+1)
    adjust_d(heap_temp,1,c_index)
    c_index -=1
  print('\n\n. ID number.  ')
  print('***********************')

  if op=='1':
    for c_index in range(1,last_index+1):
      print(' ',heap_temp[c_index])
  elif op=='2':
    c_index=last_index
    while c_index >0:
      print(' ',heap_temp[c_index])
      c_index -=1
  print('***********************')
  print(' Total member:',last_index,'\n')

def adjust_u(temp,index):
  while(index >1):
    if temp[index]<=temp[index//2]:
      break
    else:
      exchange(temp,index,index//2)
    index//=2

def adjust_d(temp,index1,index2):
  id_temp=temp[index1]
  index_temp=index1*2

  while index_temp <= index2:
    if index_temp<index2 and temp[index_temp]< temp[index_temp+1]:
      index_temp +=1
    if id_temp >=temp[index_temp]:
      break
    else:
      temp[index_temp//2]=temp[index_temp]
      index_temp *=2
  temp[index_temp//2]=id_temp

def exchange(arr,id1,id2):
  id_temp=arr[id1]
  arr[id1]=arr[id2]
  arr[id2]=id_temp

def search(id_temp):
  for c_index in range(1,len(heap_tree)):
    if id_temp==heap_tree[c_index]:
      return c_index
  return 0


def main():
  while True:
    print('\n*************HeapTree Program**********')
    print('            <1> login                ')
    print('            <2> logout                ')
    print('            <3> Show                ')
    print('            <4> Exit                ')
    print('\n**********************************')

    try:
      option=input('         Choice:        ')
    except ValueError:
      print('Not a correct number.')
      print('Try again\n')

    if option=='1':
      insert_f()
    elif option=='2':
      delete_f()
    elif option=='3':
      display_f()
    elif option=='4':
      break
      #sys.exit(0)
    else:
      print('  Invalid option!!  ')

main()
```
